# String

````python
"""
This script demonstrates basic string handling in Python using the variable `name`.

Functionality:
1. Assigns a string value to the variable `name`.
2. Prints the value of `name`.
3. Prints the type of the variable `name`.
4. Prints the memory address (ID) of the variable `name`.
5. Lists all non-dunder (non "__") attributes and methods associated with the string object.
"""

name = "Mark"

print(name)  # Print the string value
print(type(name))  # Display the type of the variable (should be <class 'str'>)
print(id(name))  # Show the memory address where the string is stored
print([i for i in dir(name) if "__" not in i])  # List of string methods and attributes excluding magic methods
````

````python
"""
This script demonstrates Python's dynamic typing and variable reassignment.

Steps:
1. Assigns a string value "Mark" to the variable `name`.
2. Reassigns the variable `name` to an integer value 700.
3. Prints the final value of `name`.
4. Prints the type of the variable after reassignment.
5. Prints the memory address (ID) of the current value stored in `name`.

This highlights that Python allows changing the type of a variable at runtime.
"""

name = "Mark"     # Initial assignment (string)
name = 700        # Reassigned to an integer

print(name)        # Output: 700
print(type(name))  # Output: <class 'int'>
print(id(name))    # Physical memory address (unique ID) of the integer object
````

````python
"""
This script assigns a string value to a variable and prints it.

Steps:
1. Assigns the string "Mark" to the variable `name`.
2. Prints the value stored in `name`.
"""

name = "Mark"
print(name)
````

````python
"""
This script demonstrates Python's use of type hints and dynamic typing.

Steps:
1. Assigns the integer value 700 to the variable `name`, with a type hint suggesting it should be a string.
2. Prints the value of `name`.

Note:
- The type hint `name: str` suggests `name` should be a string, but Python does not enforce type hints at runtime.
- Therefore, `name` will still hold the integer `700` without any error.
"""

name: str = 700
print(name)
````

````python
"""
This script demonstrates:
1. Type hinting in Python using `name: str`.
2. Printing the value, type, and memory address of a variable.
3. Listing all non-dunder (non-special) string methods and attributes.

Note:
- The type hint `name: str` indicates the variable is expected to be a string.
- Python does not enforce type hints at runtime—they're mainly for documentation and static analysis.
"""

name: str = "Mark"  # Assign a string value with a type hint

print(name)  # Print the actual value of the variable
print(type(name))  # Show the data type (should be <class 'str'>)
print(id(name))  # Print the memory address (unique object ID)
print([i for i in dir(name) if "__" not in i])  # List string methods/attributes excluding dunder methods
````

````python
"""
Title: String Method Examples on 'Ada Lovelace'

Overview:
This script demonstrates the usage of various built-in Python string methods
on the variable `name = "Ada Lovelace"`. These methods include formatting,
searching, character checking, case transformations, and whitespace handling.

Purpose:
- To serve as a hands-on reference for Python string method behaviors.
- To clarify which methods take arguments and which do not.
- To illustrate how each method works through direct `print()` examples.

Important Notes:
- Strings in Python are immutable, so all transformation methods return a new string.
- Some methods return a Boolean or numeric result based on condition checks.
- Methods that can raise exceptions (e.g., `.index()`) are used with safe input.

Demonstrated Method Categories:
- Case conversion: upper(), lower(), title(), capitalize(), swapcase(), casefold()
- Search: find(), index(), rfind(), rindex(), startswith(), endswith(), count()
- Formatting: center(), ljust(), rjust(), zfill(), format(), format_map()
- Stripping and splitting: strip(), lstrip(), rstrip(), split(), rsplit(), partition(), rpartition(), splitlines()
- Checking content: isalpha(), isdigit(), isdecimal(), isnumeric(), isalnum(), isascii(), isprintable(), isspace(), isidentifier(), istitle(), isupper(), islower()
- Modification: replace(), translate(), removeprefix(), removesuffix(), join(), maketrans()
- Encoding: encode(), expandtabs()

Each method is used in a print statement with a short example and visible output.
"""


name: str = "Ada Lovelace"

print("capitalize:", name.capitalize())
print("casefold:", name.casefold())
print("center:", name.center(20, '-'))
print("count('a'):", name.count('a'))
print("encode:", name.encode())
print("endswith('lace'):", name.endswith('lace'))
print("expandtabs:", "A\tB\tC".expandtabs(4))
print("find('Love'):", name.find('Love'))
print("format:", "My name is {}".format(name))
print("format_map:", "{f} {l}".format_map({'f': 'Ada', 'l': 'Lovelace'}))
print("index('Love'):", name.index('Love'))
print("isalnum (\"Ada123\"):", "Ada123".isalnum())
print("isalpha (\"Ada\"):", "Ada".isalpha())
print("isascii:", name.isascii())
print("isdecimal (\"123\"):", "123".isdecimal())
print("isdigit (\"123\"):", "123".isdigit())
print("isidentifier (\"name_1\"):", "name_1".isidentifier())
print("islower:", name.islower())
print("isnumeric (\"123\"):", "123".isnumeric())
print("isprintable:", name.isprintable())
print("isspace (\"   \"):", "   ".isspace())
print("istitle:", name.istitle())
print("isupper:", name.isupper())
print("join:", ",".join(['Ada', 'Lovelace']))
print("ljust:", name.ljust(20, '.'))
print("lower:", name.lower())
print("lstrip:", "   Ada".lstrip())
print("maketrans:", str.maketrans("A", "@"))
print("partition(' '):", name.partition(' '))
print("removeprefix('Ada '):", name.removeprefix('Ada '))
print("removesuffix('lace'):", name.removesuffix('lace'))
print("replace('Ada', 'Lady'):", name.replace('Ada', 'Lady'))
print("rfind('a'):", name.rfind('a'))
print("rindex('a'):", name.rindex('a'))
print("rjust:", name.rjust(20, '.'))
print("rpartition(' '):", name.rpartition(' '))
print("rsplit():", name.rsplit())
print("rstrip:", "Ada   ".rstrip())
print("split():", name.split())
print("splitlines():", "Line1\nLine2".splitlines())
print("startswith('Ada'):", name.startswith('Ada'))
print("strip:", "  Ada  ".strip())
print("swapcase:", name.swapcase())
print("title:", name.title())
print("translate:", name.translate(str.maketrans("A", "@")))
print("upper:", name.upper())
print("zfill('42'):", "42".zfill(5))
````

````python
"""
Title: Understanding Python Variables, Type Hints, and Introspection

Objective:
This tutorial demonstrates how to:
1. Declare a variable with a type hint.
2. Print the variable's value.
3. Inspect the type and memory address of the variable.
4. Explore the available methods and attributes of the variable using introspection.

Concepts Covered:
- Variable assignment
- Type hinting (`name: int`)
- Built-in functions: `print()`, `type()`, `id()`, `dir()`
- List comprehensions for filtering dunder (double underscore) methods

Note:
Although we specify `name: int`, Python does not enforce type hints at runtime. 
They are primarily used by IDEs, linters, and type checkers like `mypy`.
"""

# Step 1: Assign an integer value to the variable with a type hint
name: int = 2000

# Step 2: Print the value of the variable
print(name)  # Output: 2000

# Step 3: Print the data type of the variable
print(type(name))  # Output: <class 'int'>

# Step 4: Print the memory address (unique ID) of the variable's value
print(id(name))  # Output: (varies by system and execution)

# Step 5: List available methods and attributes of the integer object, excluding special (dunder) methods
print([i for i in dir(name) if "__" not in i])
````

````python
"""
Title: Python Variable Declaration with Float Type Hint and Object Introspection

Objective:
This tutorial demonstrates how to:
1. Declare a variable with a float type hint.
2. Print the variable's value.
3. Check the type and memory address of the variable.
4. List the available methods and attributes of a float object using introspection.

Concepts Covered:
- Variable declaration with type hints (`name: float`)
- Runtime type and memory inspection using built-in functions
- Introspection with `dir()` and filtering out special (dunder) methods

Note:
Python type hints (e.g., `name: float`) are for documentation and static analysis purposes only. 
They do not enforce the variable's type at runtime.
"""

# Step 1: Assign a float value to the variable using a type hint
name: float = 7.0

# Step 2: Print the value of the variable
print(name)  # Output: 7.0

# Step 3: Print the type of the variable
print(type(name))  # Output: <class 'float'>

# Step 4: Print the memory address (unique ID) of the float object
print(id(name))  # Output: (varies per execution)

# Step 5: Print all non-special (non-dunder) methods and attributes of the float object
print([i for i in dir(name) if "__" not in i])
# Sample Output: ['as_integer_ratio', 'conjugate', 'fromhex', 'hex', 'imag', 'is_integer', 'real']
````

````python
"""
Title: Python List Declaration with Type Hint and Introspection

Objective:
This tutorial demonstrates how to:
1. Declare a list of strings using type hinting (`list[str]`).
2. Print the contents of the list.
3. Inspect the list's type and memory address.
4. Explore the list object's available methods and attributes using introspection.

Concepts Covered:
- List declaration with type hints
- Type checking and memory inspection using `type()` and `id()`
- Filtering non-special methods from `dir()` for list objects

Note:
- The type hint `list[str]` suggests that the list should contain only string elements.
- Python does not enforce type hints at runtime—they're for developer clarity and tooling support.
"""

# Step 1: Declare a list of strings with a type hint
name: list[str] = ['a', 'b', 'c']

# Step 2: Print the list
print(name)  # Output: ['a', 'b', 'c']

# Step 3: Print the type of the variable
print(type(name))  # Output: <class 'list'>

# Step 4: Print the memory address (ID) of the list object
print(id(name))  # Output: (varies each time)

# Step 5: Display non-dunder (non "__") methods and attributes of the list object
print([i for i in dir(name) if "__" not in i])
# Sample Output: ['append', 'clear', 'copy', 'count', 'extend', 'index', 'insert', 
#                 'pop', 'remove', 'reverse', 'sort']
````

````python
"""
Title: Python Tuple Declaration with Type Hint and Introspection

Objective:
This tutorial demonstrates how to:
1. Declare a tuple with multiple types using Python type hints (`tuple[str, int, float]`).
2. Print the tuple's contents.
3. Display the data type and memory address of the tuple.
4. List available methods and attributes of a tuple object using introspection.

Concepts Covered:
- Tuple declaration and immutability
- Type hinting with heterogeneous types
- Built-in functions: `print()`, `type()`, `id()`, and `dir()`
- Filtering non-dunder (non-special) methods for clarity

Note:
- The type hint `tuple[str, int, float]` means the tuple is expected to contain:
  - A string in the first position
  - An integer in the second
  - A float in the third
- Type hints are not enforced at runtime in Python.
"""

# Step 1: Declare a tuple with specific types using a type hint
name: tuple[str, int, float] = ('pakistan', 7, 2.0)

# Step 2: Print the tuple
print(name)  # Output: ('pakistan', 7, 2.0)

# Step 3: Print the type of the variable
print(type(name))  # Output: <class 'tuple'>

# Step 4: Print the memory address (object ID) of the tuple
print(id(name))  # Output: (varies by execution)

# Step 5: List all non-dunder methods and attributes of the tuple
print([i for i in dir(name) if "__" not in i])
# Sample Output: ['count', 'index']
````

````python
"""
Title: Python Tuple Type Hint vs. Runtime Behavior (With Mismatched Types)

Objective:
This tutorial demonstrates:
1. Declaring a tuple with an explicit type hint (`tuple[str, int, float]`).
2. Assigning a tuple that **does not conform** to the specified type hint.
3. Exploring what happens at runtime.
4. Using introspection to examine the object.

Key Lesson:
- Type hints in Python (e.g., `tuple[str, int, float]`) suggest expected types for static analysis tools like `mypy`, but **are not enforced at runtime**.
- This means that even if the third element is a `str` instead of the expected `float`, Python will still run the program without raising a runtime error.

Expected vs Actual:
- Expected: `('pakistan', 7, 2.0)` → (str, int, float)
- Actual:   `('pakistan', 7, "aaa")` → (str, int, str)
"""

# Step 1: Declare a tuple with a mismatched type assignment
name: tuple[str, int, float] = ('pakistan', 7, "aaa")  # 'aaa' is a str, not a float

# Step 2: Print the value of the tuple
print(name)  # Output: ('pakistan', 7, 'aaa')

# Step 3: Print the actual type of the object
print(type(name))  # Output: <class 'tuple'>

# Step 4: Print the memory address (object ID)
print(id(name))  # Output: (varies at runtime)

# Step 5: List non-dunder methods and attributes
print([i for i in dir(name) if "__" not in i])
# Sample Output: ['count', 'index']
````

````python
"""
Title: Using `Any` Type Hint in Python and Exploring Runtime Object Introspection

Objective:
This tutorial demonstrates:
1. Declaring a variable with the `Any` type from the `typing` module.
2. Assigning a string value to the variable.
3. Printing the value, type, and memory address.
4. Listing all non-dunder (non-special) methods and attributes of the object.

Concepts Covered:
- Type hinting with `Any` from `typing`
- Dynamic typing and introspection in Python
- The flexibility and risks of using `Any` in type-annotated code

Note:
- `Any` is a special type hint indicating that a variable can hold *any* type of value.
- Python does not enforce type hints at runtime.
- Useful when the type is unknown or intentionally flexible, but reduces type safety.
"""

from typing import Any

# Step 1: Declare a variable with a flexible type hint (Any)
name: Any = "Pakistan"  # This could later be reassigned to any type

# Step 2: Print the current value of the variable
print(name)  # Output: Pakistan

# Step 3: Print the actual data type of the object
print(type(name))  # Output: <class 'str'>

# Step 4: Print the memory address (ID) of the object
print(id(name))  # Output: (varies each time the script runs)

# Step 5: List all non-dunder methods and attributes of the object
print([i for i in dir(name) if "__" not in i])
# Output (sample): ['capitalize', 'casefold', 'center', ..., 'zfill']
````

## PEMDAS

````python
print(3 + 2 - 2 * 4 /2 + 2)
#     3+2−2×4÷2+2
#     3 + 2 - 4 + 2
#      6    - 8
````

Alright, let's solve the equation step by step. We'll use the order of operations, often remembered by the acronym PEMDAS/BODMAS:

- **P**arentheses/**B**rackets
- **E**xponents (ie Powers and Square Roots, etc.)/**O**rders (ie Powers and Square Roots, etc.)
- **MD** Multiplication and Division (left-to-right)
- **AS** Addition and Subtraction (left-to-right)

Given the expression:

\[ 3 + 2 - 2 \times 4 \div 2 + 2 \]

Let's solve it step by step:

1. **Multiplication and Division**:
\[ 2 \times 4 = 8 \]
\[ 8 \div 2 = 4 \]

Replacing these values in the expression, we get:

\[ 3 + 2 - 4 + 2 \]

2. **Addition and Subtraction** (from left to right):
\[ 3 + 2 = 5 \]
\[ 5 - 4 = 1 \]
\[ 1 + 2 = 3 \]

So, the final answer is:

\[ 3 \]

````python
from typing import Union
PerType = Union[float, int]

percentages : list[PerType] = [88, 99.9, 50, 51,65,70]

grades : list[str] = []

for per in percentages:
    grade : str = ""

    if (per >= 0) and (per < 33):
        grade = "Fail"
    elif (per >= 33) and (per < 40):
        grade = "E"
    elif (per >= 40) and (per < 50):
        grade = "D"
    elif (per >= 50) and (per < 60):
        grade = "C"
    elif (per >= 60) and (per <70) :
        grade = "B"
    elif (per >= 70) and (per <80) :
        grade = "A"
    elif (per >=80) and (per <= 100):
        grade = "A+"

    grades.append(grade)

print(percentages)
print(grades)

    




````

````python
from typing import Callable

def my_decorator(func: Callable[[], None])-> Callable[[], None]:
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
````

````python
from typing import Callable

def my_decorator(func: Callable[[int], None]) -> Callable[[int], None]:
    def wrapper(num1: int) -> None:
        print("Something is happening before the function is called.")
        func(num1)
        print("Something is happening after the function is called.")
    return wrapper

@my_decorator
def say_hello(num1: int) -> None:
    print(num1)

say_hello(100)
````

````python
def factorial(x: int) -> int:
    """This is a recursive function
    to find the factorial of an integer"""
    if x == 1:
        return 1
    else:
        return (x * factorial(x-1))
        #       5*4*3*2*1

num = 5
print("The factorial of", num, "is", factorial(num))
````

## Method Overloading

````python

from typing import Union, overload

class Adder:
    @overload
    def add(self, x: int, y: int) -> int:
        ...
        
    @overload
    def add(self, x: float, y: float) -> float:
        ...
        
    @overload
    def add(self, x: str, y: str) -> str:
        ...
        
    
    def add(self, x: Union[int, float, str], y: Union[int, float, str]) -> Union[int, float, str]:
        if isinstance(x, int) and isinstance(y, int):
            return x + y
        elif isinstance(x, float) and isinstance(y, float):
            return x + y
        elif isinstance(x, str) and isinstance(y, str):
            return x + y
        else:
            raise TypeError("Invalid argument types!")

# Usage examples
adder = Adder()
result1 = adder.add(1, 2)  # Should return 3
result2 = adder.add(1.5, 2.5)  # Should return 4.0
result3 = adder.add("Hello, ", "world!")  # Should return "Hello, world!"

print(result1)
print(result2)
print(result3)

# Overridding & polymorphism
````

````python
import pandas as pd
import pandera as pa

# data to validate
df = pd.DataFrame({
    "column1": [1, 4, 0, 10, 9],
    "column2": [-1.3, -1.4, -2.9, -10.1, -20.4],
    "column3": ["value_1", "value_2", "value_3", "value_2", "value_1"],
})

# define schema
schema = pa.DataFrameSchema({
    "column1": pa.Column(int, checks=pa.Check.le(10)),
    "column2": pa.Column(float, checks=pa.Check.lt(-1.2)),
    "column3": pa.Column(str, checks=[
        pa.Check.str_startswith("value_"),
        # define custom checks as functions that take a series as input and
        # outputs a boolean or boolean Series
        pa.Check(lambda s: s.str.split("_", expand=True).shape[1] == 2)
    ]),
})

validated_df = schema(df)
print(validated_df)
````

````python
def fn_obtained(s1: int, s2: int, s3: int, s4: int, s5: int)->tuple[int, int,float, str]:
    total: int = 500
    obtained :int = s1 + s2 + s3 + s4 + s5
    per : float = obtained / total * 100
    grade : str = ""
    if per >= 80:
        grade =  "A+"
    elif per >= 70:
        grade = "A"
    elif per >= 60:
        grade = "B"
    elif per >= 50:
        grade = "C"
    elif per >= 40:
        grade = "D"
    elif per >= 33:
        grade = "E"
    else:
        grade = "F"
    return total, obtained, per, grade


df1[["Total",'obtained', 'percentage', "Grade"]] = df1[["s1", "s2", "s3", "s4", "s5"]].apply(lambda x:fn_obtained(*x), 
                                                                                             axis=1,
                                                                                             result_type='expand')
df1
````

In pandas, when you have a `datetime` column in a DataFrame, you can format the date and time information using the `strftime` method which allows for flexible formatting by specifying format codes. Each code represents a different element of the date and time. Below are some of the commonly used format codes that you can use with `strftime`:

- `%a` - Abbreviated weekday name.
- `%A` - Full weekday name.
- `%w` - Weekday as a decimal number, where 0 is Sunday and 6 is Saturday.
- `%d` - Day of the month as a zero-padded decimal number.
- `%b` - Abbreviated month name.
- `%B` - Full month name.
- `%m` - Month as a zero-padded decimal number.
- `%y` - Year without century as a zero-padded decimal number.
- `%Y` - Year with century as a decimal number.
- `%H` - Hour (24-hour clock) as a zero-padded decimal number.
- `%I` - Hour (12-hour clock) as a zero-padded decimal number.
- `%p` - Locale’s equivalent of either AM or PM.
- `%M` - Minute as a zero-padded decimal number.
- `%S` - Second as a zero-padded decimal number.
- `%f` - Microsecond as a decimal number, zero-padded on the left.
- `%z` - UTC offset in the form ±HHMM[SS[.ffffff]] (empty string if the object is naive).
- `%Z` - Time zone name (empty string if the object is naive).
- `%j` - Day of the year as a zero-padded decimal number.
- `%U` - Week number of the year (Sunday as the first day of the week) as a zero-padded decimal number.
- `%W` - Week number of the year (Monday as the first day of the week) as a zero-padded decimal number.
- `%c` - Locale’s appropriate date and time representation.
- `%x` - Locale’s appropriate date representation.
- `%X` - Locale’s appropriate time representation.
- `%G` - ISO 8601 year with century representing the year that contains the greater part of the ISO week (`%V`).
- `%u` - ISO 8601 weekday as a decimal number where 1 is Monday.
- `%V` - ISO 8601 week number as a decimal number with Monday as the first day of the week.

Here's a small example of how to use `strftime` with a pandas DataFrame:

```python
import pandas as pd

# Create a sample DataFrame with datetime column
df = pd.DataFrame({
    'datetime': pd.date_range(start='2023-01-01', periods=3, freq='D')
})

# Format the 'datetime' column as a string with a specific format
df['formatted_date'] = df['datetime'].dt.strftime('%Y-%m-%d %H:%M:%S')

print(df)
```

The `'formatted_date'` column will contain strings representing the dates in the format specified by the `strftime` method. You can combine any of the symbols above to create the date format string as per your requirements.
